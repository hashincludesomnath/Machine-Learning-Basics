```python
# import datascience libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load Dataset**

```python
dataset = pd.read_csv('Grade_Set_2.csv')
dataset.head()
```

**Extracting independent and dependent variables**

```python
X = dataset.iloc[:,0:1].values # independent variable
y = dataset.iloc[:,-1].values # dependent variable
```

**Visualize data**

```python
plt.style.use('fivethirtyeight')
```

```python
plt.scatter(X,y)
plt.xlabel('Hours Studied') , plt.ylabel('Grade')
plt.title('Hours vs Grade')
plt.show()
```

**Build a Linear Regression model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lin = LinearRegression() # class
model_lin.fit(X,y) # traning model
y_pred_lin = model_lin.predict(X) # evaluating model
dataset['pred_linear_reg'] = y_pred_lin # append into my dataset
dataset
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred_lin,'r')
plt.xlabel('Hours Studied') , plt.ylabel('Grade')
plt.title('Hours vs Grade')
plt.legend(['Linear Regression'])
plt.show()
```

```python
X_new = np.concatenate((X,X**2,X**3),axis = 1) # appending polynomial features
```

**Building Polynomial Regression**
- degree = 3

```python
model_poly2 = LinearRegression() # linear regression
```

```python
model_poly2.fit(X_new,y) # fit or traning model
y_pred_poly2 = model_poly2.predict(X_new) # evaluating
dataset['Predict_poly2 '] = y_pred_poly2
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred_lin,'r') # linear regression
plt.plot(X,y_pred_poly2,'g') # ploynomial regression
plt.xlabel('Hours Studied') , plt.ylabel('Grade')
plt.title('Hours vs Grade')
plt.legend(['Linear Regression','Polynomial Linear Regression'])
plt.show()
```
