```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
meal = [1,2,3,4,5,6]
tip = [5,17,11,8,14,5]

x = np.array(meal)
y = np.array(tip)

plt.style.use('fivethirtyeight')
plt.subplot(2,2,1)
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('Meal no.')
plt.ylabel('Tip Amount')

plt.subplot(2,2,2)
plt.boxplot(x)
plt.xlabel('Meal No.')
plt.ylabel('Tip Amount')
plt.title('X Boxplot')

plt.subplot(2,2,3)
plt.boxplot(y)
plt.xlabel('Meal No.')
plt.ylabel('Tip Amount')
```

```python
x = x.reshape(-1,1) # Coloumn Matrix
# x = x.reshape(1,-1) Row matrix
```

```python
x.shape
```

```python
y = y.reshape(-1,1) # Coloumn Matrix
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
# sklearn -> class
# linear_model -> class
# LinearRegression -> class
```

```python
model = LinearRegression() # Machine Learning Model - Linear
```

```python
# Training Machine Learning Model
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.xlabel('Meal no.')
plt.ylabel('Tip Amount')
plt.legend('Linear Regression')
```
